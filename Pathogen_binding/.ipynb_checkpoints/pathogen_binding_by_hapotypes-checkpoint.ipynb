```r
# Process column V1 to split by "." and combine the first and second parts
###read in antigens and their peptides list
peptide_pathogens <- read.delim("antigens_", sep = "[", header = FALSE)

# Process column V1: First split by ".", then split the resulting parts by "_"
peptide_pathogens$V1 <- sapply(peptide_pathogens$V1, function(x) {
  parts_dot <- strsplit(x, split = "\\.")[[1]]  # Split by "."
  parts_underscore <- strsplit(parts_dot[1], split = "_")[[1]]  # Split first part by "_"
  paste(parts_underscore[1], parts_underscore[2], sep = "_")  # Combine the first two parts with "_"
})
peptide_pathogens$antigen <- gsub(pattern = ">", replacement = "", peptide_pathogens$V1)
peptide_pathogens$antigen <- gsub(pattern = ">", replacement = "", peptide_pathogens$V1)


peptide_pathogens$antigen <- gsub(pattern = "_NA", replacement = "", peptide_pathogens$antigen)
peptide_pathogens$pathogen <- sapply(peptide_pathogens$V2, function(x) {
  parts <- strsplit(x, split = " ")[[1]]  # Split by space
  paste(parts[1], parts[2], sep = " ")   # Combine the first two parts with a space
})


peptide_pathogens$pathogen <- gsub("]", "",  peptide_pathogens$pathogen)
peptide_pathogens$pathogen <- gsub(" NA", "",  peptide_pathogens$pathogen)

peptide_pathogens[peptide_pathogens$pathogen == "Human gammaherpesvirus","pathogen"] <- "Herpes"
peptide_pathogens[peptide_pathogens$pathogen == "Human herpesvirus","pathogen"] <- "Herpes"
peptide_pathogens[peptide_pathogens$pathogen == "Giardia lamblia","pathogen"] <- "Giardia"
peptide_pathogens[peptide_pathogens$pathogen == "Giardia intestinalis","pathogen"] <- "Giardia"

library(dplyr)

# For each all_SB_* data frame, join with peptide_pathogens to add the corresponding pathogen
all_SB_A <- all_SB_A %>%
  left_join(peptide_pathogens, by = c("antigen" = "antigen")) %>%
  mutate(pathogen = coalesce(pathogen, "Unknown")) # Fills unmatched antigens with "Unknown"

all_SB_B <- all_SB_B %>%
  left_join(peptide_pathogens, by = c("antigen" = "antigen")) %>%
  mutate(pathogen = coalesce(pathogen, "Unknown"))

all_SB_C <- all_SB_C %>%
  left_join(peptide_pathogens, by = c("antigen" = "antigen")) %>%
  mutate(pathogen = coalesce(pathogen, "Unknown"))

all_SB_DQ <- all_SB_DQ %>%
  left_join(peptide_pathogens, by = c("antigen" = "antigen")) %>%
  mutate(pathogen = coalesce(pathogen, "Unknown"))

all_SB_DRB1 <- all_SB_DRB1 %>%
  left_join(peptide_pathogens, by = c("antigen" = "antigen")) %>%
  mutate(pathogen = coalesce(pathogen, "Unknown"))  



length(unique(peptide_pathogens$pathogen))
head(peptide_pathogens)
head(all_SB_DRB1)
```

```r
test_hap <- as.character(allele_freq_df[1,1])
test_hap
strsplit(test_hap, split="-")[[1]][[1]]
###look up number of unique peptide that can bind the haplotype per pathogen
look_up <- function(hap) {
    A <- strsplit(hap, split = "-")[[1]][[1]]
    B <- strsplit(hap, split = "-")[[1]][[2]]
    C <- strsplit(hap, split = "-")[[1]][[3]]
    DQ <- paste0(strsplit(hap, split = "-")[[1]][[5]], "-", strsplit(hap, split = "-")[[1]][[6]])
    DQ <- gsub(pattern = ":", replacement = "", DQ)
    DRB1 <- strsplit(hap, split = "-")[[1]][[7]]
    DRB1 <- gsub(pattern = ":", replacement = "", DRB1)

    bound <- c()
    pathogens <- data.frame(peptide = character(), antigen = character(), pathogen = character())  # Initialize empty dataframe with specific column names
    
    for (locus in c("A", "B", "C", "DQ", "DRB1")) {
        binding <- get(paste0("all_SB_", locus))
        loc <- get(locus)
        #print(loc)
        # Subset the relevant columns and ensure that 'binding' has matching columns
        binding_subset <- binding[binding$allele %in% loc, c("V3", "antigen", "pathogen")]
        
        #print(head(binding_subset))
        # Use rbind to combine the data frames
        ###stores the df with the peptide, antigen, and pathogen
        pathogens <- rbind(pathogens, binding_subset)
        #print(head(binding[binding$allele %in% loc, 3]))
        # Store the bound peptides
        bound <- c(bound, binding[binding$allele %in% loc, 3])
    }
  
    # Print the pathogens data frame
    ## get only the unique peptides (V3) from the pathogends df
    pathogens <- pathogens[!duplicated(pathogens$V3),]
    pathogens$pathogen <- gsub("]", "",  pathogens$pathogen)
    pathogens$pathogen <- gsub(" NA", "",  pathogens$pathogen)

    # tabulate the unique pathogen peptides per pathogen
    pathogen_table <- table(pathogens$pathogen)  # You can change 'pathogen' if needed
    # get total unique peptides bound
    unique_bound_length <- length(unique(bound))
    #print(pathogen_table)
    #print(bound)
    return(list(pathogen_table = pathogen_table, unique_bound_length = unique_bound_length))
}
look_up(test_hap)
```

```r
results <- apply(allele_freq_df, 1, function(row) look_up(row["Allele"]))

# Extract pathogen table frequencies and unique bound lengths
allele_freq_df$pathogen_frequencies <- lapply(results, function(res) res$pathogen_table)
allele_freq_df$unique_bound_length <- sapply(results, function(res) res$unique_bound_length)

# If you want to expand the pathogen frequencies into separate columns:
pathogen_names <- unique(unlist(lapply(results, function(res) names(res$pathogen_table))))
for (pathogen in pathogen_names) {
  allele_freq_df[[paste0("freq_", pathogen)]] <- sapply(
    allele_freq_df$pathogen_frequencies,
    function(table) ifelse(pathogen %in% names(table), table[pathogen], 0)
  )
}

# Drop the intermediate pathogen_frequencies column if no longer needed
allele_freq_df$pathogen_frequencies <- NULL
```

```r
allele_freq_df$Frequency <- as.numeric(allele_freq_df$Frequency)

# Check pathogen frequency columns and convert to numeric if needed
pathogen_columns <- grep("^freq_", names(allele_freq_df), value = TRUE)
allele_freq_df[pathogen_columns] <- lapply(allele_freq_df[pathogen_columns], as.numeric)

# Initialize a list to store model results
model_results <- list()

# Fit a linear model for each pathogen frequency column
for (pathogen in pathogen_columns) {
  model <- lm(Frequency ~ get(pathogen), data = allele_freq_df)
  model_results[[pathogen]] <- summary(model)
}


summary_table <- data.frame(
  Pathogen = gsub("^freq_", "", pathogen_columns),
  Estimate = sapply(model_results, function(m) coef(m)["get(pathogen)", "Estimate"]),
  P_value = sapply(model_results, function(m) coef(m)["get(pathogen)", "Pr(>|t|)"]),
  R_squared = sapply(model_results, function(m) m$r.squared)
)



summary_table$Pathogen <- gsub("]", "", summary_table$Pathogen)
summary_table$Pathogen <- gsub(" NA", "", summary_table$Pathogen)

# Display summary table
print(summary_table)
```

```r
library(ggplot2)
library(rlang)
library(cowplot)

# Create a list to store the individual plots
plot_list <- list()
options(repr.plot.width = 30, repr.plot.height = 30)
pathogens <- unique(summary_table$Pathogen)
# Generate scatter plots and store them in the list
for (pathogen in pathogens) {
  # Extract corresponding values from the summary table
  pathogen_name <- pathogen
  p_value <- summary_table[summary_table$Pathogen == pathogen_name,]$P_value
  
  p <- ggplot(allele_freq_df, aes(y = .data[[paste0("freq_",pathogen)]], x = .data$Frequency)) +
    geom_point() +
    geom_smooth(method = "lm", color = "blue") +
    labs(
      title = paste("Haplotype freq vs", pathogen_name, "peptides bound"),
      y = paste("Number of unique peptides"),
      x = "Haplotype Frequency"
    ) +
    annotate(
      "text",
      x = Inf, y = Inf,
      label = paste0("p = ", format(p_value, digits = 3)),
      hjust = 1.1, vjust = 2,
      size = 5, color = "red"
    ) +
theme_minimal() +
  theme(
    legend.position = "none", 
    axis.text.x = element_text(size = 20),    # Increase x-axis label size
    axis.text.y = element_text(size = 20),    # Increase y-axis label size
    # axis.ticks.x = element_line(size = 3),   # Increase x-axis tick size
    # axis.ticks.y = element_line(size = 3),   # Increase y-axis tick size
    plot.title = element_text(size = 12, face = "bold", hjust = 0.5),  # Increase title size and make it bold
    panel.background = element_blank(),  # Remove the gray background
    plot.background = element_blank(),  # Remove the overall background
    axis.title.x = element_text(size = 20),  # Increase x-axis title size
    axis.title.y = element_text(size = 20)   # Increase y-axis title size
  )
  
  plot_list[[pathogen]] <- p
}

# Split the plots into multiple grids of 3x3
n_plots <- length(plot_list)
n_per_grid <- 25  # 3x3 grid
n_grids <- ceiling(n_plots / n_per_grid)

# Plot each grid
for (i in seq_len(n_grids)) {
  start_idx <- (i - 1) * n_per_grid + 1
  end_idx <- min(i * n_per_grid, n_plots)
  
  grid_plots <- plot_list[start_idx:end_idx]
  
  # Use cowplot::plot_grid() to arrange plots in a grid
  grid <- plot_grid(plotlist = grid_plots, ncol = 5, nrow = 5)
  print(grid)
}
```
